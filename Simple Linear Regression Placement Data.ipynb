```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
df=pd.read_csv('d:/dataset/placement.csv')
df.head()
```

```python
df.info()
```

```python
X=df.iloc[:, :-1].values  
y=df.iloc[:, 1].values
```

```python
print(X)
```

```python
print(y)
```

```python
plt.scatter(X,y)
plt.show()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=1/5,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
print("X_train:",X_train)
print()
print("X_test:",X_test)
print()
print("y_train:",y_train)
print()
print("y_test:",y_test)
```

```python
lrm=LinearRegression()
lrm.fit(X_train,y_train)
```

```python
m=lrm.coef_[0]
c=lrm.intercept_

print(f"m value is {m}")
print(f"c value is {c}")
```

```python
y_pred=lrm.predict(X_test)
print(y_pred)
```

```python
df1 = pd.DataFrame(y_pred, columns=['Predicted Package'])
df2 = pd.DataFrame(y_test,columns=['Original Package'])
```

```python
pd.concat([df1,df2],axis=1).head()
```

```python
plt.scatter(X_train, y_train, color='b') 
plt.scatter(X_train, lrm.predict(X_train), color='c') 
plt.title('CGPA vs Package(Training set)')
plt.xlabel('CGPA')
plt.ylabel('Package')
plt.show()
```

```python
plt.scatter(X_test, y_test, color='b')
plt.scatter(X_test, lrm.predict(X_test), color='c')
plt.title('CGPA vs Package(Test set)')
plt.xlabel('CGPA')
plt.ylabel('Package')
plt.show()
```

```python
R2=metrics.r2_score(y_test,y_pred)
MSE=metrics.mean_squared_error(y_test,y_pred)
MAE=metrics.mean_absolute_error(y_test,y_pred)
RMSE=np.sqrt(metrics.mean_squared_error(y_test,y_pred))

print("R2 = ",R2)
print("MSE = ",MSE)
print("MAE =",MAE)
print("RMSE =",RMSE)
```
