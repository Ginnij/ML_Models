```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn

import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv('d:/dataset/airline.csv')
df.head()
```

```python
df['Arrival Delay in Minutes']=df['Arrival Delay in Minutes'].fillna(0)
```

```python
Customer_Type=pd.get_dummies(df['Customer_Type'], drop_first=True)
```

```python
Type_of_Travel=pd.get_dummies(df['Type_of_Travel'],drop_first=True)
```

```python
Class=pd.get_dummies(df['Class'],drop_first=True)
```

```python
satisfaction=pd.get_dummies(df['satisfaction'],drop_first=True)
```

```python
df.drop(['Customer_Type', 'Type_of_Travel', 'Class','satisfaction'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df=pd.concat([df,Customer_Type,Type_of_Travel,Class,satisfaction],axis=1)
```

```python
df.head()
```

```python
sn.pairplot(df)
plt.show()
```

```python
X=df.iloc[:,:-1].values
```

```python
X
```

```python
y=df.iloc[:,-1].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train,y_train)
```

```python
predictions=log_reg.predict(X_test)
print(predictions)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn import metrics
```

```python
metrics.roc_curve(y_test,predictions)
```

```python
metrics.roc_auc_score(y_test,predictions)
```

```python
metrics.accuracy_score(y_test,predictions)
```
